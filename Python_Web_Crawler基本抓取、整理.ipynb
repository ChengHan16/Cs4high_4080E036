```
import requests
from bs4 import BeautifulSoup
import json
url = "https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=ssd&page=1&sort=sale/dc"
res = requests.get(url)
data = json.loads(res.text)
data
webdatas = data["prods"]
len (webdatas)

for product in webdatas : 
  print(product['name'])
  print(product['price'])
```
